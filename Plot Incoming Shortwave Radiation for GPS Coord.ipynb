```python
import numpy as np
from math import cos
import matplotlib.pyplot as plt

Iinitial= 1367
lat = 43.00 *(np.pi/180)
long = 87.9 *(np.pi/180)
n = np.arange(1,365,1)
theta = ((2*np.pi*n)/365)
distRatio = (1.00011 + .034221*np.cos(theta) + .001280*np.sin(theta) + 
            .000719*(2*(np.cos(theta-1))**2) + .000077*(2*(np.sin(theta)*np.cos(theta))))

I = Iinitial*distRatio
t = 15.5
h = ((np.pi * t)/12) - long
decl= (.409)* np.cos((2*np.pi*(n-173))/365.25)

cosZ = np.sin(lat)*np.sin(decl) - np.cos(lat)*np.cos(decl)*cos(h)
print(cosZ)
Kex= I*cosZ
print(Kex)

```

```python
plt.plot(n,Kex, label="Solar Radiation at Top of Atmopshere",color ='blue')
plt.plot(n,Kex*.90, label="Incoming Shortwave Radiation", color ='teal')
plt.plot(n,Kex*.90*.85 , label="Max Shortwave Radiation Collected", color ='red')
plt.xlabel("Day of Year in 2020")
plt.ylabel("Incoming Shortwave Radiation W/m^2")
plt.title("Daily Incoming Shortwave Radiation for Milwaukee, WI at 1600Z")
plt.grid(color='grey', linestyle=':', linewidth=1)
plt.legend()

#max_y = np.max(Kex)  # Find the maximum y value
#max_x = x[Kex.argmax()]  # Find the x value corresponding to the maximum y value
#min_y = np.min(Kex)  # Find the maximum y value
#min_x = x[Kex.argmin()] 
#print (max_x,'DOY,', max_y,'Extrarerrestrial radiation')
#print (min_x,'DOY,', min_y,'Extrarerrestrial radiation')
```

```python
Iinitial= 1367 #solar constant
lat = 39.7392 #Lat for Denver CO
lon = 1.83 #lon for denver
n1 = 271   #DOY of Sept 27th,2020

theta = ((2*np.pi*n1)/365)  
distRatio = (1.00011 + .034221*np.cos(theta) + .001280*np.sin(theta) + 
            .000719*(2*(np.cos(theta-1))**2) + .000077*(2*(np.sin(theta)*np.cos(theta)))) #sun-earth distance eq

I = Iinitial*distRatio  #Extraterrestrial radiation normal to beam

t= np.arange(0,23,1)   #time -24 hour day, measured every 30 minutes
h= (((np.pi)*t)/12) - lon  #The hour angle 
decl= (np.deg2rad(23.5)* np.cos((2*np.pi*(n1-173))/365.25)) #Angle of declination
cosZ = np.sin(np.deg2rad(lat))*np.sin(decl)+ np.cos(np.deg2rad(lat))*np.cos(decl)*np.cos(h) #Solar Zenith Angle 

print("decl",decl)

print('hour',h)
print('sinB',cosZ)

new_Z = []

for val in cosZ:
    if val >= 0:
        new_Z.append(val)
    else:
        new_Z.append(0)

Kex= np.multiply(new_Z,I)
print(Kex)

plt.plot(t,Kex)
plt.xlabel("Hour of Day")
plt.grid(color='grey', linestyle=':', linewidth=1)
plt.ylabel("Extraterrestrial Radiation W/m^2")
plt.title("Diurnal Variation in Extraterrestrial Radiation for Denver,CO \n for September 27th, 2020")
```
